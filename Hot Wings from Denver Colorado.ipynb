```python
import pandas as pd
import json 
from yelpapi import YelpAPI
with open('C:/Users/Sean/.edit/yelp_api.json') as f:
    login = json.load(f)
login.keys()
```

```python
yelp_api = YelpAPI(login['api-key'], timeout_s = 3.0)
yelp_api
```

```python
hot_wings = yelp_api.search_query(location = 'Denver, CO',
                                 term = 'Buffalo Wings')
type(hot_wings)
hot_wings.keys()
```

```python
hot_wings0 = pd.DataFrame(hot_wings['businesses'])
hot_wings0.head(3)
```

```python
hot_wings['total']
```

```python
# gonna be alot of pages, will limit it to 20
hot_wings20 = yelp_api.search_query(location = 'Denver, CO',
                                   term = 'Hot Wings',
                                   offset = 20)
```

```python
hot_wings20 = pd.DataFrame(hot_wings20['businesses'])
hot_wings20.head(3)
```

```python
#concatinating pages
all_wings = pd.concat([hot_wings0,hot_wings20],
                     ignore_index=True)
display(all_wings.head(3),all_wings.tail(3))
```
